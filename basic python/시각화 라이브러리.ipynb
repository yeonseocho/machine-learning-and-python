```python
import matplotlib.pyplot as plt

#
%matplotlib inline
```

```python
#도화지를 깔고
plt.figure()
#해당 리스트를 기준으로 하는 선을 그립니다.
plt.plot([1,2,4,8,16])
#보여주기
plt.show()
```

```python
#example
x=['Americano', 'Cafe Latte', "Vanilla Latte",'Cold Brew', "Flat White"]
y=[1463, 301, 866, 905, 274]
```

## 도화지에 그림을 그려봅시다.
### 순서대로 하나씩 세팅해봅니다.
+ figure&figsize
+ barplot
+ grid
+ title
+ axis label
+ lineplot
+ set label
+ legend
+ set limit in axis

```python
plt.figure(figsize=(8,6)) #도화지 크기
#figure아래에 componenet들을 추가를 해야 plt 에 대한 내용으로 이해한다.
plt.grid(alpha=0.3, color="black", linewidth=0.5)
plt.bar(x,y)
#barh -horizontal line
plt.show()
```
