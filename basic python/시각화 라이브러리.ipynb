## 도화지에 그림을 그려봅시다.
### 순서대로 하나씩 세팅해봅니다.
+ figure&figsize
+ barplot
+ grid
+ title
+ axis label
+ lineplot
+ set label
+ legend
+ set limit in axis

```python
plt.figure(figsize=(8,6)) #도화지 크기
#figure아래에 componenet들을 추가를 해야 plt 에 대한 내용으로 이해한다.
plt.grid(alpha=0.3, color="black", linewidth=0.5)
plt.title("Coffe Sales in March", fontsize=20, loc="left")
plt.xlabel("Menu", fontsize=12)
plt.ylabel("Sales", fontsize=12)
#plt.plot(x,y) #선
plt.bar(x,y, color="lightcoral", label="Sales") #바
plt.plot(sorted([100,200,400,800,1600], reverse=True), label="Google Trend")
plt.legend(loc="best") # 각 그래프가 지칭하는 것 명시
#barh -horizontal line
#plt.ylim(0, 2000)
plt.yticks([num for num in range(0, 1700, 100)])
plt.show()
```

```python
#scatterplot
import numpy as np

x=np.random.random(100)
y=np.random.random(100)

plt.figure(figsize=(6,6))
plt.scatter(x,y,color="green", marker="^", s=15, alpha=0.3) #marker: 점 찍히는 모양, marker option 확인해보기, s=size
plt.show()
```
